# Stock Data Analysis Plan

**Dataset Columns:** `symbol`, `date`, `close`, `high`, `low`, `open`, `volume`

---

## 1. Problem Statement
**Goal:** Help investors and portfolio managers make informed buy, hold, or sell decisions by analyzing historical stock performance and identifying trends, volatility, and trading opportunities.

---

## 2. Key Questions to Answer

### Price Performance & Trends
- How has the stock price changed over time (daily, weekly, monthly)?
- Are there long-term upward or downward trends?

### Volatility & Risk
- How volatile is the stock over a given period?
- What is the range between daily highs and lows?

### Trading Activity
- Are there periods of unusually high trading volume?
- Is there a correlation between volume spikes and price movement?

### Opportunities & Patterns
- Are there recurring patterns in the stock price (seasonality, trends)?
- When would have been the optimal times to buy or sell historically?

---

## 3. Metric KPIs (Numerical)

| KPI | Description | Calculation |
|-----|------------|-------------|
| Average Closing Price | Central tendency of stock price | Mean of `close` prices |
| Price Volatility | Measures risk | Standard deviation of `close` prices |
| Average Daily Range | Indicates intraday price fluctuation | Average of `high - low` |
| Average Trading Volume | Measures market activity and liquidity | Mean of `volume` |

---

## 4. Non-Metric KPIs (Categorical / Event-Based)

| KPI | Description |
|-----|------------|
| Upward Trend Periods | Number of consecutive days/weeks/months with closing price > previous |
| Downward Trend Periods | Number of consecutive days/weeks/months with closing price < previous |
| High-Volume Spikes | Count of days where volume exceeds threshold (e.g., 2Ã— average volume) |
| Price Breakouts | Days where closing price exceeds previous high (resistance levels) |
| Gap Days | Days with large opening price gaps relative to previous close |

---

## 5. Expected Business Impact
- **Performance Monitoring:** Track trends, highs, and lows over time
- **Risk Assessment:** Identify volatile periods and mitigate potential losses
- **Trading Opportunities:** Pinpoint buy/sell signals through breakouts and gaps
- **Portfolio Optimization:** Use metrics to decide which stocks to hold or divest
